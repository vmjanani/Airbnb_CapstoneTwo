```python
#there are 135 rows where beds empty , so removing those rows
airbnb_df['beds'].isna()
```

```python
#drop the columns where there are no beds
airbnb_df.dropna(subset=['beds'],inplace=True)
```

```python
#there are still around 980 rooms which have beds and even in ammenities there are ammenities but the bedroom is NaN
airbnb_df[airbnb_df['bedrooms'].isna()]
```

```python
#assign bedroom , as beds/2 and round to the next higest integer
airbnb_df['bedrooms']=airbnb_df['bedrooms'].fillna(airbnb_df['beds']/2)
airbnb_df['bedrooms']=np.ceil(airbnb_df['bedrooms'])
```

```python
airbnb_df.info()
```

```python
#convert the bedrooma and beds as int
airbnb_df['bedrooms']=airbnb_df['bedrooms'].astype('int')
```

```python
airbnb_df['beds']=airbnb_df['beds'].astype('int')
```

```python
airbnb_df.info()
```

```python
#removed the $ and the , from the values 
airbnb_df['price']=airbnb_df['price'].str.replace('$','').str.replace(',','')
```

```python
#convert the price type as float
airbnb_df['price']=airbnb_df['price'].astype('float')
```
